```python
import tweepy
import pandas as pd
import re
import nltk
import string
```

```python
consumerKey="your consumer key"
consumerSecret="your consumer secret key"
accessToken="your access Token"
accessTokenSecret="your access Token Secret"
```

```python
auth=tweepy.OAuthHandler(consumerKey,consumerSecret)
auth.set_access_token(accessToken,accessTokenSecret)
api=tweepy.API(auth,wait_on_rate_limit = True)
```

```python
search_word = ["Fan2Play"]
#date_until = "2021-04-17"
date_since = "2021-04-16"
tweets = tweepy.Cursor(api.search,q=search_word,lang='en',since=date_since).items(1000)
#tweet=[[tweet.text] for tweet in tweets]
#pd.set_option('display.max_colwidth',100)    
tweet_df = pd.DataFrame([[tweet.text,tweet.user.location] for tweet in tweets],columns=['tweet','location'])
print(tweet_df)
```

```python
#create a function to clean tweets
def cleanTxt(text):
  text = re.sub('@[A-Za-z0-9]+', '', text) # Removing @mention
  text = re.sub('#', '', text) # Removing hashtag
  text = re.sub('RT[\s]+', '', text) # Removing RT
  text = re.sub('https[?:\/\/\sA-Za-z0-9]+', '', text) # Removing hyperlink
  #text = re.sub(':', '', text)
  text = "".join([char.lower() for char in text if char not in string.punctuation]) # Removing punctuation and normalization
  return text

# Clean the tweets
tweet_df['tweet_clean'] = tweet_df['tweet'].apply(cleanTxt)

# Show the cleaned tweets
tweet_df
```

```python
## Tokenization 
def tokenize(text):
  tokens = re.split('\W+',text)
  return tokens
tweet_df['tweet_clean']=tweet_df['tweet_clean'].apply(tokenize)  
tweet_df
#tweeText = tweet_df['tweet'] 
#from nltk.tokenize import  word_tokenize
#nltk.download('stopwords')
#from nltk.corpus import stopwords
#stop_words = stopwords.words('english')
#stop_words.append('play')
#tweet=TweetTokenizer()
#tweet.tokenize(tweet_df['tweet'])
#tweet_Tok = tweeText.apply(word_tokenize)
#tweet_Tok
#tweet_Tok_stop = [i for i in tweet_Tok if not i in stop_words]
#tweet_Tok_stop 
```

```python
## Remove stop words
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  txt_clean = [word for word in text if word not in stopwords]
  return txt_clean

tweet_df['tweet_clean'] = tweet_df['tweet_clean'].apply(remove_stopwords)
tweet_df
```

```python
## stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stemming(text):
  stem_text = [ps.stem(word) for word in text]
  return stem_text

tweet_df['tweet_clean'] = tweet_df['tweet_clean'].apply(stemming)
tweet_df  
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
```

```python
tweet_df['tweet_clean']=tweet_df['tweet_clean'].apply(lambda x: ' '.join(map(str, x)))
tweet_df['score']=tweet_df['tweet_clean'].apply(lambda tweet: sid.polarity_scores(tweet))
tweet_df.head
```

```python
tweet_df['compound']=tweet_df['score'].apply(lambda score_dict: score_dict['compound'])
tweet_df.head()
```

```python
def getAnalysis(score):
  
  if score < 0:
   return 'Negative'
  elif score == 0:
   return 'Neutral'
  else:
   return 'Positive'

tweet_df['Analysis_2'] = tweet_df['compound'].apply(getAnalysis)
tweet_df.head()
```
