# Hypothesis Testing

The Chi-Square test of independence is used to determine if there is a significant relationship 
between two nominal (categorical) variables.<br>
The frequency of each category for one nominal variable is compared across the categories of 
the second nominal variable.<br>
The data can be displayed in a contingency table where each row represents a category for one 
variable and each column represents a category for the other variable.<br>



H0: There is no significance relationship between Creator and type.<br>
Vs<br>
H1: There is significance relationship between Creator and type.<br>

```python
ct1=pd.crosstab(data.Type,data.creator,margins=True)
ct1
```

```python
from scipy.stats import chi2_contingency
```

```python
chival, pval, df, exp=chi2_contingency(ct1)
chival, pval, df, exp
```

```python
import researchpy 
```

```python
ct1,res,exp = researchpy.crosstab(data['creator'],data['Type'],test='chi-square',expected_freqs=True)
```

```python
res
```

#### From the above table P-Value are indicate that there is significant relationship between creator and type

H0: There is no significance relationship between Creator and Sentiment.<br>
